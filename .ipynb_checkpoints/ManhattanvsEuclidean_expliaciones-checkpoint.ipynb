```python
#En este notebook vemos como de diferentes son las explicaciones dadas por lime con la distancia euclidea y con la distancia manhattan
#para ello siempre daremos la explicaciones con la mitad  de los atributos disponibles y evaluaremos en que proporcion los atributos 
#escogidos por las distancias son los mismos o no.

#Antes de realizar el experimento lo que espero es que cuando la dimensionalidad es baja las explicaciones de ambas distancias se parezcan mas 
#que cuando la dimensionalidad es alta, debo encontrar una medida justa que reduzca el efecto de la dimensionalidad y solo valore como de diferentes 
#son las explicaciones

#Dada una instancia i, y un datset con n atributos, realizamos la expliacion de lime con n/2 atributos con la distancia euclidea y la distancia 
#manhattan.
#la medida de similitud sera S = m/(n/2) = m*2/n con m el numero de atributos que coinciden en ambas expliaciones, S es un valor  entre  0 y 1, 
#cuanto mas cercano sea a 1 mas se parecen las explicaciones. 
#En nuestro experimento calcularemos el valor S sobre todas las intancias del test set y haremos la media. Despues representaremos estas medias 
#con respecto a la dimensionalidad 

#Obs. Con 2 atributos, la expliacion se hace con 1, con una eleccion de atributos aleatoria, la probabilidad de S = 1 es 0.5
#     Con 3 atributos, la explicacon se hace  con 3/2 --> 1, la prob de S = 1 es 0.33
#     Con 4 atributos , la expliacacion se hace on 2, la prob  de S = 1 es 1/6 (puedes hacer 6 parejas distintas con 4 atributos disponibles)
#     Con k atributos, la expliacion se hace ccon [k/2], la prob de S = 1 es 1/Comb(k,[k/2])

#Para que la medida sea realmente justa entonces debemos ajustar de la siguiente manera (no dependa del numero de atributos)

# S_relativ = m/(n/2) * Comb(k,[k/2]), ahora S_relativ in [0,Comb(k,[K/2])] (Esta medida no es  la que uso realmente despues)
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
import random
import warnings
warnings.filterwarnings('ignore')
from sklearn import ensemble
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

import lime.lime_tabular
from lime_stability.stability import LimeTabularExplainerOvr

seed = 4
random.seed(seed)
```

```python
#la funcion explication_study la cambio para que nos devuelva tambien los atributos que se usan en la explicacion.
def Explication_study(expl,instance,num_features,num_samples,distance):
  

  exp = expl.explain_instance(X_test[instance],classifier.predict_proba, 
                                      num_features=num_features,num_samples=num_samples,
                                      distance_metric=distance)
  used_features = list(zip(*exp.local_exp[1]))[0]
  
  csi,vsi = expl.check_stability(X_test[instance],classifier.predict_proba, 
                                          num_features=num_features,num_samples=num_samples,
                                          distance_metric=distance,n_calls = 10)
  return exp.score,csi,vsi,used_features
```

```python
def get_S(used_features_eu,used_features_man,dimension):
  #Dadas dos listas de atributos usados y la dimension te devuelve la medida de similitud, no voy a usar S_relativ
  count = 0
  for f in used_features_eu:
    if f in used_features_man:
      count = count+1
  S = count*2/dimension
  S_relativ = S*math.comb(dimension,int(dimension/2))
  return S,S_relativ
```

```python
#Cambio plot_data para que grafique tambien el VSI y la similitud

def Plot_data(euclidean_data,manhattan_data,similitud_data,min_dim,max_dim,step_dim):
  l = len(euclidean_data)
  dimensiones = np.arange(min_dim,max_dim,step_dim)
  eu_R = np.zeros(l)
  eu_CSI = np.zeros(l)
  eu_VSI = np.zeros(l)
  man_R = np.zeros(l)
  man_CSI = np.zeros(l)
  man_VSI = np.zeros(l)
  similitud = np.zeros(l)
  for i in range(l):
    eu_R[i] = euclidean_data[i][0]
    eu_CSI[i] = euclidean_data[i][1]
    eu_VSI[i] = euclidean_data[i][2]
    man_R[i] = manhattan_data[i][0]
    man_CSI[i] = manhattan_data[i][1]
    man_VSI[i] = manhattan_data[i][2]
    similitud[i] = similitud_data[i]


  #Graficamos 
  fig, ax = plt.subplots(2, 2, figsize=(16,6), sharey = False)
  fig.tight_layout(pad=5.0) #espaciar las graficas

  ax[0,0].plot(dimensiones, eu_R, color='r', label='euclidean')
  ax[0,0].plot(dimensiones, man_R, color='g', label='manhattan')
  ax[0,0].set_title('R^2: Euclidean vs Manhattan')
  ax[0,0].legend()

  ax[0,1].plot(dimensiones, eu_CSI, color='r', label='euclidean')
  ax[0,1].plot(dimensiones, man_CSI, color='g', label='manhattan')
  ax[0,1].set_title('CSI: Euclidean vs Manhattan')
  ax[0,1].legend()

  ax[1,0].plot(dimensiones, eu_VSI, color='r', label='euclidean')
  ax[1,0].plot(dimensiones, man_VSI, color='g', label='manhattan')
  ax[1,0].set_title('VSI: Euclidean vs Manhattan')
  ax[1,0].legend()

  ax[1,1].plot(dimensiones, similitud, color='b')
  ax[1,1].set_title('Similarity Euclidean-Manhattan')
  ax[1,1].legend()

  plt.show()
```

```python
#Hacemos ahora el estudio de verdad
min_dimension=2
max_dimension=20
step=2
Eu_data = []
Man_data = []
S_data = [] #Aqui guardatemos los datos  de similitud

for d in range(min_dimension,max_dimension,step): #Cuantos atributos va a haber en el dataset
  X, y = make_classification(n_samples=500, n_features=d, n_informative = d, n_redundant = 0, n_classes=2, n_clusters_per_class=1, random_state=seed)
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.90, random_state=seed)
  n=X_test.shape[0]
  classifier = ensemble.RandomForestClassifier()
  classifier.fit(X_train, y_train);
  explainer_eu = LimeTabularExplainerOvr(X_train, verbose = False, discretize_continuous=True, random_state = seed) #Default kernel 
  explainer_man = LimeTabularExplainerOvr(X_train, kernel_width= d*0.75, verbose = False, discretize_continuous=True, random_state = seed) #Custom kernel
  #Euclidean calculations
  R_array = np.zeros(n) 
  CSI_array = np.zeros(n)
  VSI_array = np.zeros(n)
  features_eu = [] #Lista donde guardo los atributos usados por la euclidea en todas las instancias del test set
  for i in range(n):
    r,csi,vsi,used_features_eu = Explication_study(explainer_eu,i,int(d/2),4000,"euclidean") #Usamos la mitad de los atributos
    R_array[i]=r
    CSI_array[i]=csi
    VSI_array[i]=vsi
    features_eu.append(used_features_eu) #Aqui los guardo
  Eu_data.append((np.mean(R_array),np.mean(CSI_array),np.mean(VSI_array)))
  #Manhattan calculations
  R_array = np.zeros(n) 
  CSI_array = np.zeros(n)
  VSI_array = np.zeros(n)
  features_man = [] 
  for i in range(n):
    r,csi,vsi,used_features_man = Explication_study(explainer_man,i,int(d/2),4000,"manhattan") #Usamos la mitad de los atributos
    R_array[i]=r
    CSI_array[i]=csi
    VSI_array[i]=vsi
    features_man.append(used_features_man)
  Man_data.append((np.mean(R_array),np.mean(CSI_array),np.mean(VSI_array)))
  #Similitud calculations
  S_array = np.zeros(n) #En la posicion i del array esta la medida de similitud para la instancia i 
  for i in range(n):
    S_array[i] = get_S(features_eu[i],features_man[i],d)[0] #No tomamos el S_relativ
  S_data.append(np.mean(S_array)) #hacemos la media y la guardamos en S_data
  print(d)

Plot_data(Eu_data,Man_data,S_data,min_dimension,max_dimension,step)
```
